## import module 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## import data

```python
df = pd.read_csv("PakWheelsDataSet.csv")
df
```

## Make

```python
groupByMake = df.groupby(['Make'])['Make'].count()
groupByMake = groupByMake.to_frame()
groupByMake.columns = ['transaction']
groupByMake = groupByMake.reset_index()
groupByMake.sort_values("transaction", ascending = False, inplace=True)
groupByMake['%'] = groupByMake['transaction'] / groupByMake['transaction'].sum()*100
groupByMake
```

## Name

```python
groupByName = df.groupby(['Name', 'Make'])['Name'].count()
groupByName = groupByName.to_frame()
groupByName.columns = ['transaction']
groupByName = groupByName.reset_index()
groupByName.sort_values("transaction", ascending = False, inplace=True)
groupByName['%'] = groupByName['transaction'] / groupByName['transaction'].sum()*100
# groupByName.head(20)['%'].sum()
groupByName.head(20)
```

## Make&Price

```python
priceAndMake = df.loc[:,['Price', 'Make']]
def interval_map(x):
    if x > 10000000:
        return 'K >10M'
    elif x > 5000000:
        return 'J 5M~10M'
    elif x > 4000000:
        return 'I 4M~5M'
    elif x > 3000000:
        return 'H 3M~4M'
    elif x > 2000000:
        return 'G 2M~3M'
    elif x > 1000000:
        return 'F 1M~2M'
    elif x > 750000:
        return 'E 750K~1M'
    elif x > 500000:
        return 'D 500K~750K'
    elif x > 250000:
        return 'C 250K~500K'
    elif x > 100000:
        return 'B 100K~250K'
    else:
        return 'A <100K'

priceAndMake["PriceInterval"] = priceAndMake["Price"].map(interval_map)

groupByPriceAndMake = priceAndMake.groupby(['PriceInterval', 'Make'])['PriceInterval'].count()
groupByPriceAndMake = groupByPriceAndMake.to_frame() #把上列的Series轉成DataFrame
groupByPriceAndMake.columns = ['Count']
groupByPriceAndMake = groupByPriceAndMake.reset_index()
groupByPriceAndMake.sort_values("PriceInterval", ascending = True, inplace=True)
groupByPriceAndMake.head(20)

groupByPriceAndMake  = groupByPriceAndMake.pivot(index = 'Make',columns = 'PriceInterval',values = 'Count')
groupByPriceAndMake.columns = ['250K~500K', '500K~750K', '750K~1M', '1M~2M', '2M~3M', '3M~4M', '4M~5M', '5M~10M', '>10M']

sns.set(font_scale = 2)
plt.figure(figsize=(20,30))
heatmap = sns.heatmap(groupByPriceAndMake, cmap = 'rocket_r') #cmap:制定熱力圖顏色
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 30) #繪製x軸標示角度
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation = 0) #繪製y軸標示角度
plt.xlabel('PriceInterval')
plt.ylabel('Make')
plt.show()
```

## Make&City

```python
downSize = df.loc[:,['City']]
downSize = downSize.groupby(['City'])['City'].count()
downSize = downSize.to_frame()
downSize.columns = ['Count']
downSize = downSize.reset_index()
mask = downSize['Count'] > 500 #這裡決定減少的城市數量
downSize = downSize[mask]
downSize = downSize['City'].to_list()
len(downSize)
```

```python
makeAndCity = df.loc[:,['Make', 'City']]
groupByMakeAndCity= makeAndCity.groupby(['Make', 'City'])['Make'].count()
groupByMakeAndCity = groupByMakeAndCity.to_frame()
groupByMakeAndCity.columns = ['Count']
groupByMakeAndCity = groupByMakeAndCity.reset_index()
groupByMakeAndCity.sort_values('City',  ascending = True, inplace=True)
mask = groupByMakeAndCity['City'].isin(downSize)
groupByMakeAndCity = groupByMakeAndCity[mask]
groupByMakeAndCity  = groupByMakeAndCity.pivot(index = 'Make',columns = 'City',values = 'Count')
groupByMakeAndCity

sns.set(font_scale = 2)
plt.figure(figsize=(40,30))
heatmap = sns.heatmap(groupByMakeAndCity, cmap = 'rocket_r') #cmap:制定熱力圖顏色
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 30) #繪製x軸標示角度
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation = 0) #繪製y軸標示角度
plt.xlabel('City')
plt.ylabel('Make')
plt.show()
```
